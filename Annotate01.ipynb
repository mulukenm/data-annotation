```python
import os
from pigeon import annotate
from IPython.display import display, Image
```

```python
folder_path = '/content/Boots and Sneakers Dataset'
image_names = os.listdir(folder_path)
image_paths = [os.path.join(folder_path, image_name) for image_name in image_names]
```

# New Section

```python
image_paths
```

```python
len(image_paths)
```

```python
lable_options = ['boot', 'sneakers']
```

```python
annotations = annotate(image_paths, options=lable_options, display_fn=lambda filename:display(Image(filename)))
```

```python
annotations
```

```python
import pandas as pd
```

```python
df_annotation = pd.DataFrame(annotations, columns=['filename', 'label'])

df_annotation
```

```python
df_annotation.to_csv('boots_and_sneakers_annotations.csv', index=False)
```
